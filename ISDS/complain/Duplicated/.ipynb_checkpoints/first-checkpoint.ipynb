```python
import pandas as pd
import numpy as np
from difflib import SequenceMatcher
import matplotlib.pyplot as plt
import math
from konlpy.tag import Okt
from tqdm import tqdm, trange
```

```python
df = pd.read_csv("duplicted_freq.csv")
```

```python
df['dup_distribution'] = df.total_freq - df.daily_freq
```

```python
df['dup_distribution'].describe()
```

```python
df['ratio'] = df['dup_distribution'] / df['total_freq']
df_ratio_underhalf = df[df['ratio'] < 0.5]
df_ratio_under90 = df[df['ratio'] < 0.9]
```

```python
plt.subplot(222)
plt.scatter(df['total_freq'], df['daily_freq'])
plt.scatter(df_ratio_underhalf['total_freq'], df_ratio_underhalf['daily_freq'])
#plt.scatter(df_ratio_under90['total_freq'], df_ratio_under90['daily_freq'])
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Total_frequncy (log)')
plt.ylabel('Daily_frequncy (log)')
plt.subplot(221)
plt.scatter(df['total_freq'], df['daily_freq'])
plt.scatter(df_ratio_underhalf['total_freq'], df_ratio_underhalf['daily_freq'])
#plt.scatter(df_ratio_under90['total_freq'], df_ratio_under90['daily_freq'])
plt.xlabel('Total_frequncy')
plt.ylabel('Daily_frequncy')
plt.show()
```

```python
okt = Okt()

deleted = 0
def Tokenize(x):
    if len(x) > 3000: 
        global deleted 
        deleted += 1
        return ['TOO', "LONG", 'TXT']
    
    else:
        return okt.nouns(x)
```

```python
df['tokens'] = list(map(Tokenize, df['본문']))
```
