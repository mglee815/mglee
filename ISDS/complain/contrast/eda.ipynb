```python
import pandas as pd
```

```python
petitions_tf = pd.read_pickle("petitions_tf.pkl")
complain_tf = pd.read_pickle("complain_tf_py.pkl")
```

```python
def Calculate_ratio(df):
    df = pd.DataFrame(list(df.items()), columns=['word', 'freq'])
    df.sort_values(by = ['freq'], axis = 0, ascending = False, inplace= True)
    total = sum(df['freq'])
    temp = list(map(lambda x : x / total, df.freq))
    df['ratio'] = temp
    return df
```

```python
pet_tf_df = Calculate_ratio(petitions_tf)
comp_tf_df = Calculate_ratio(complain_tf)
```

```python
pet_tf_df
comp_tf_df
```

```python
pet_com = pd.merge(pet_tf_df, comp_tf_df, how = 'inner', on ='word')
```

```python
pet_com['pet / comp'] = pet_com['ratio_x'] / pet_com['ratio_y']
pet_com.sort_values(by = ['pet / comp'], axis = 0, ascending = False, inplace= True)
pet_com.to_csv("petition_complain_contrast.csv")
```
