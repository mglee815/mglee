```python
from konlpy.tag import Okt
okt = Okt()
test = [
    "이것은 메켑 테스트 입니다.",
    "이것도 메켑 메켑 테스트 임",
    "그것은 메켑 테스트 테스트 입니까?"
]
for txt in test:
    print(okt.nouns(txt))
```

```python
from tqdm import tqdm 
def tf_count(lst):
    word_dict = {}
    for txt in tqdm(lst):
        txt = str(txt)
        if len(txt) < 1000:
            lst = okt.nouns(txt)
        else:
            lst = ['TOO', "LONG"]
        for word in lst:
            try:
                word_dict[word] = word_dict[word] + 1 
            except:
                word_dict[word] = 1
    return word_dict
```

```python
petitions_tf = tf_count(petitions_body)
```

```python
pd.to_pickle(petitions_tf, "petitions_tf.pkl")
```

```python
complain_tf = tf_count(complain_body)
pd.to_pickle(complain_tf, "complain_tf.pkl")
```
