```python
import pandas as pd
```

```python
body = pd.read_excel("/home/mglee/VSCODE/git/ISDS/region/body0118.xlsx")
id = pd.read_excel("/home/mglee/VSCODE/git/ISDS/region/id.xlsx")
```

```python
body[10:60]
```

```python
temp = pd.merge(body, id[['광역', '자치']], how = 'left', left_on = 'dist', right_on = '자치')
```

```python
temp[10:20]
```

```python
temp['Unnamed: 0'] = temp['광역']
temp.drop(['광역', '자치'], axis = 1, inplace = True)
temp[40:60]
```

```python
temp.to_excel("/home/mglee/VSCODE/git/ISDS/region/merge0118.xlsx")
```

```python
j_rank = temp[temp['dist'] == '제주시'][['순위:보건', '순위:복지', '순위:교육', '순위:문화']]
s_rank = temp[temp['dist'] == '서귀포시'][['순위:보건', '순위:복지', '순위:교육', '순위:문화']]
```

```python
j_rank['순위:보건'].item()
```

```python
check_lst = ['순위:보건', '순위:복지', '순위:교육', '순위:문화']

global_lst = []

for col in check_lst:
    lst = []
    for item_rank in temp[col]:
        modi = 0
        if item_rank > j_rank[col].item():
            modi += 1
        if item_rank > s_rank[col].item():
            modi += 1
        lst.append(item_rank - modi)
    global_lst.append(lst)
```

```python
new_rank = pd.DataFrame(global_lst).T

new_df = pd.concat([temp, new_rank], axis = 1)
new_df.to_excel("/home/mglee/VSCODE/git/ISDS/region/temp.xlsx")
```
